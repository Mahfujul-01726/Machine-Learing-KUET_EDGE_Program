# 1. What is Feature Selection?
Feature selection is the process of selecting the most important features (independent variables) from a dataset that contribute significantly to the prediction of the target variable.

**Goal:**
* Reduce dimensionality.
* Remove irrelevant or redundant features.
* Improve model performance.

# 2. Why Perform Feature Selection?

1. Reduces Overfitting: Removing irrelevant features reduces the chance of the model learning noise.
2. Improves Accuracy: Only the relevant features help in better prediction.
3. Decreases Training Time: Fewer features mean faster training and simpler models.

# 3. Types of Feature Selection Methods

1. Filter Methods: Based on statistical tests.
2. Wrapper Methods: Based on model performance (e.g., Recursive Feature Elimination).
3. Embedded Methods: Feature selection performed during model training (e.g., LASSO).

# 4. Filter-Based Feature Selection

**Steps:**
1. Apply statistical metrics to filter features.
2. Remove features that don't meet the criteria.

# 5. Duplicate Features

```python
import pandas as pd

# Create a dataset with duplicate features
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [1, 2, 3, 4, 5],  # Duplicate of column A
    'C': [5, 4, 3, 2, 1]
}

df = pd.DataFrame(data)

# Identify duplicate columns
duplicate_columns = df.T.duplicated().tolist()
df_cleaned = df.loc[:, ~df.T.duplicated()]
print("Cleaned DataFrame:\n", df_cleaned)
```

# 6. Variance Threshold

Screenshot 2025-01-09 160148.png

```python
from sklearn.feature_selection import VarianceThreshold

# Create a dataset with low variance
data = [[0, 2, 0],
        [0, 3, 0],
        [0, 4, 0]]

# Apply variance threshold
selector = VarianceThreshold(threshold=0.1)
selected_data = selector.fit_transform(data)
print("Selected Features based on Variance:\n", selected_data)
```

# 7. Correlation-Based Feature Selection

Screenshot 2025-01-09 160245.png

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Generate correlated data
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [1, 2, 3, 4, 5]
})

# Compute correlation matrix
corr_matrix = df.corr()

# Plot heatmap
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

# 8. ANOVA(Analysis of Variance)

Screenshot 2025-01-09 160511.png

```python
from sklearn.feature_selection import f_classif
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)
f_values, p_values = f_classif(X, y)
print("F-values:\n", f_values)
print("P-values:\n", p_values)
```

# 9. Chi-Square Test ( Test)

Screenshot 2025-01-09 160725.png

```python
from sklearn.feature_selection import chi2
import numpy as np

X = np.array([[1, 0, 1],
              [2, 1, 1],
              [0, 0, 0],
              [1, 1, 1]])
y = np.array([0, 1, 0, 1])

chi_scores, p_values = chi2(X, y)
print("Chi-Square Scores:\n", chi_scores)
print("P-values:\n", p_values)
```

# 10. Advantages and Disadvantages of Feature Selection:
*  Advantages:
1. Reduces overfitting.
2. Improves model accuracy and interpretability.
3. Decreases training time.
*  Disadvantages:
1. Risk of removing informative features.
2. Some methods assume feature independence (e.g., Chi-Square).
