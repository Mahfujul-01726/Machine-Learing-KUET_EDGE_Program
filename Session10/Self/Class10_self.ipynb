# 2. Plotting Simple Functions , Labels, Legends , Multiple Plots

Plotting Simple Function

```python
import numpy as np

# Function to plot: y = x^2
x = np.linspace(-10, 10, 100)
y = x ** 2

plt.plot(x, y, label="y = x^2")  # Add a label for the legend
plt.title("Plot of y = x^2")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()  # Display the legend
plt.show()
```

Adding Labels and Legends

```python
# Adding labels and legends
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin(x)", color="blue")
plt.plot(x, y2, label="cos(x)", color="red")
plt.title("Sine and Cosine Waves")
plt.xlabel("X-axis (radians)")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

# Multiple Plots

```python
# Multiple plots in a single figure
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(2, 1, 1)  # 2 rows, 1 column, first plot
plt.plot(x, y1, color="green")
plt.title("Sine Wave")

plt.subplot(2, 1, 2)  # 2 rows, 1 column, second plot
plt.plot(x, y2, color="orange")
plt.title("Cosine Wave")

plt.tight_layout()  # Adjust spacing
plt.show()
```

# 3. About scatter plots

```python
# Scatter Plot
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

plt.scatter(x, y, color='purple')
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# 4. Bar charts

Example: Vertical Bar Chart

```python
# Bar Chart
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 8, 5]

plt.bar(categories, values, color='cyan')
plt.title("Vertical Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

Horizontal Bar Chart

```python
# Horizontal Bar Chart
plt.barh(categories, values, color='magenta')
plt.title("Horizontal Bar Chart")
plt.xlabel("Values")
plt.ylabel("Categories")
plt.show()
```

# 5. Histogram

```python
# Histogram
data = [7, 8, 5, 5, 6, 8, 9, 6, 5, 6, 7, 9, 10, 6, 7]

plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.show()
```

# 6. Pie Chart

```python
# Pie Chart
labels = ['A', 'B', 'C', 'D']
sizes = [30, 40, 20, 10]
colors = ['gold', 'blue', 'lightcoral', 'green']
explode = (0.1, 0, 0, 0)  # Highlight the first slice

plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart Example")
plt.show()
```

# 7. Changing Styles of Plots

```python
# List of available styles
print(plt.style.available)

# Using a specific style
plt.style.use('ggplot')

# Plot with style
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label="sin(x)")
plt.title("Styled Plot with ggplot")
plt.legend()
plt.show()
```
