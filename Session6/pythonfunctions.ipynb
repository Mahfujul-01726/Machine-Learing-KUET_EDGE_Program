The train_linear_regression_model function encapsulates the training process for a linear regression model.
Nested functions calculate_weights, predict, and mean_squared_error are defined within train_linear_regression_model to handle specific tasks related to training, prediction, and error calculation.
Each nested function performs a specific calculation or operation, making the code modular and easier to understand.
The nested functions have access to the variables defined in the outer function, allowing them to operate on the input data and intermediate results.

The expression weights = np.linalg.inv(X.T @ X) @ X.T @ y is a common calculation in linear regression where the weights (coefficients) of the model are computed using the normal equation. Let's break down this calculation step by step:

X.T @ X: This part calculates the dot product of the transpose of the feature matrix X and X itself. This operation is often performed to compute the covariance matrix of the features.
np.linalg.inv(X.T @ X): This part calculates the inverse of the matrix resulting from the dot product X.T @ X. In linear regression, this inverse is used to find the "optimal" weights that minimize the sum of squared errors.
X.T @ y: This part calculates the dot product of the transpose of the feature matrix X and the target variable y. This operation essentially computes the covariance between the features and the target.
Combining the above steps:
np.linalg.inv(X.T @ X) @ X.T @ y: After computing the inverse of X.T @ X, the result is multiplied by the transpose of X and then by y. This operation yields the optimal weights for the linear regression model.
In summary, the expression weights = np.linalg.inv(X.T @ X) @ X.T @ y is used in linear regression to calculate the optimal weights that minimize the squared error between the predicted values (X @ weights) and the actual target values y. This formula is derived from the normal equation method in linear regression and provides a closed-form solution for finding the best-fitting linear model.

```python
def train_linear_regression_model(X, y):
    def calculate_weights(X, y):
        # Calculate the weights using the normal equation
        weights = np.linalg.inv(X.T @ X) @ X.T @ y
        return weights

    def predict(X, weights):
        # Make predictions using the calculated weights
        predictions = X @ weights
        return predictions

    def mean_squared_error(y_true, y_pred):
        # Calculate the mean squared error
        mse = np.mean((y_true - y_pred) ** 2)
        return mse

    # Calculate weights
    weights = calculate_weights(X, y)

    # Make predictions
    predictions = predict(X, weights)

    # Calculate mean squared error
    mse = mean_squared_error(y, predictions)

    return weights, predictions, mse

# Sample data
import numpy as np
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Training the linear regression model
import numpy as np
weights, predictions, mse = train_linear_regression_model(X, y)

print("Weights:", weights)
print("Mean Squared Error:", mse)
```

```python
# some more functions
def is_prime(n):
    if n in [2, 3]:
        return True
    if (n == 1) or (n % 2 == 0):
        return False
    r = 3
    while r * r <= n:
        if n % r == 0:
            return False
        r += 2
    return True
print(is_prime(73), is_prime(78))
```
