# write a program to choose if the number is between 10 and 50

```python

number = float(input("Enter a number: "))

if 10 <= number <= 50:
    print("The number is between 10 and 50.")
else:
    print("The number is not between 10 and 50.")
```

# Write a program that check if the entered number is divided by both 3 and 5.

```python

number = int(input("Enter a number: "))

if number % 3 == 0 and number % 5 == 0:
    print(f"{number} is divisible by both 3 and 5.")
else:
    print(f"{number} is not divisible by both 3 and 5.")
```

# Kuet grading system

```python
def get_grade(mark):
    if mark >= 80:
        return 'A+', 4.00
    elif 75 <= mark < 80:
        return 'A', 3.75
    elif 70 <= mark < 75:
        return 'A-', 3.50
    elif 65 <= mark < 70:
        return 'B+', 3.25
    elif 60 <= mark < 65:
        return 'B', 3.00
    elif 55 <= mark < 60:
        return 'B-', 2.75
    elif 50 <= mark < 55:
        return 'C+', 2.50
    elif 45 <= mark < 50:
        return 'C', 2.25
    elif 40 <= mark < 45:
        return 'D', 2.00
    else:
        return 'F', 0.00

mark = float(input("Enter the mark: "))
grade, grade_point = get_grade(mark)
print(f"Grade: {grade}, Grade Point: {grade_point}")
```

# break continue and pass

```python
for i in range(1, 11):
    if i == 3:
        print(f"Skipping {i} with continue")
        continue
    elif i == 6:
        print(f"Stopping the loop at {i} with break")
        break
    elif i == 8:
        print(f"Placeholder action for {i} with pass")
        pass
    else:
        print(f"Processing {i}")

print("Loop finished")
```

# Task: Stress Problem


```python
# List of stress values
stress_values = [100, 250, 350, 600, 450, 500, 200]

# Loop through the list and check the stress levels
for stress in stress_values:
    if stress < 300:
        print(f"Stress {stress}: Safe")
    elif 300 <= stress < 500:
        print(f"Stress {stress}: Warning")
    else:
        print(f"Stress {stress}: Failure")
```

# Zip function in python

```python
names = ['John', 'Alice', 'Bob', 'Lucy']
scores = [85, 90, 78, 92]
ages = [25, 30, 22, 28]

res = zip(names, scores, ages)
print(list(res))
```
