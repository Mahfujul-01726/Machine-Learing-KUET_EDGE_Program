# 2. Code for MLR using Normal Equation

```python
import numpy as np

# Generate synthetic dataset
np.random.seed(42)
X1 = np.random.rand(100)
X2 = np.random.rand(100)
y = 3 + 5*X1 + 2*X2 + np.random.randn(100) * 0.5

# Combine features and add bias term
X = np.c_[np.ones(X1.shape[0]), X1, X2]

# Normal Equation to solve for beta coefficients
beta = np.linalg.inv(X.T @ X) @ X.T @ y
print("Estimated Coefficients:", beta)
```

# 5. Minimizing Error using Normal Equation

```python
# Predictions using the calculated coefficients
y_pred = X @ beta

# Mean Squared Error Calculation
mse = np.mean((y - y_pred)**2)
print("Mean Squared Error:", mse)
```

# 6. Error Function Continued (Gradient Descent Approach)

Screenshot 2025-01-08 224812.png

```python
def gradient_descent(X, y, learning_rate=0.1, epochs=1000):
    m, n = X.shape
    beta_gd = np.zeros(n)
    for _ in range(epochs):
        gradients = -(1/m) * (X.T @ (y - X @ beta_gd))
        beta_gd -= learning_rate * gradients
    return beta_gd

# Training with Gradient Descent
beta_gd = gradient_descent(X, y)
print("Beta from Gradient Descent:", beta_gd)
```
